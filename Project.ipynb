Data Normalization and data type conversion

```python
#Min-Max Normalization

df['GPA_MinMax'] = (df['GPA'] - df['GPA'].min()) / (df['GPA'].max() - df['GPA'].min())
print("\nMin-Max Normalized GPA")
print(df[['Student_ID','GPA','GPA_MinMax']])
```

```python
#Normalize Multiple Columns
cols = ['Attendance','GPA']
df_norm = (df[cols] - df[cols].min()) / (df[cols].max() - df[cols].min())
print("\nNormalized Dataset")
print(df_norm)
```

```python
# Z-score Normalization
df['GPA_Zscore'] = (df['GPA'] - df['GPA'].mean()) / df['GPA'].std()
print("\nZ-score Normalized GPA")
print(df[['Student_ID','GPA','GPA_Zscore']])
```

```python
# Decimal Scaling
df['GPA_DS'] = df['GPA'] / 100000
print("\nDecimal Scaled GPA")
print(df[['GPA','GPA_DS']])
```

Charts and Visual Encoding

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.preprocessing import LabelEncoder
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import warnings

warnings.filterwarnings('ignore')

# Load data
df = pd.read_csv('/content/raw data students.csv')

# Basic data validation
print("Checking missing values...")
print(df.isnull().sum())
df = df.dropna(subset=['GPA', 'Attendance'])  # Drop rows with missing critical values
df = df[(df['GPA'] >= 0) & (df['GPA'] <= 10)]  # Ensure GPA is valid
df = df[(df['Attendance'] >= 0) & (df['Attendance'] <= 100)]  # Ensure Attendance is valid

# Set styling
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette("husl")

print("Student Performance Analytics Dashboard")
print("Loading data...")

# Data preparation
dept_means = df.groupby('Department')['GPA'].mean().reset_index()
df['GPA_Category'] = pd.cut(df['GPA'], bins=[0,7,8,9,10], labels=['Poor','Average','Good','Excellent'])

# Encode Department for 3D plot
if 'Department_Enc' not in df.columns:
    le = LabelEncoder()
    df['Department_Enc'] = le.fit_transform(df['Department'])

# Main dashboard: 2x3 subplot
print("Creating main dashboard...")

fig = make_subplots(
    rows=2, cols=3,
    subplot_titles=('Gender Distribution', 'Department GPA Ranking', 'GPA Distribution',
                   'Attendance vs GPA', 'GPA by Living', 'Correlation Matrix'),
    specs=[[{"type": "bar"}, {"type": "bar"}, {"type": "histogram"}],
           [{"type": "scatter"}, {"type": "box"}, {"type": "heatmap"}]],
    horizontal_spacing=0.05, vertical_spacing=0.08
)

# 1. Gender distribution
gender_counts = df['Gender'].value_counts()
fig.add_trace(
    go.Bar(x=gender_counts.index, y=gender_counts.values,
           marker_color=['#FF6B6B', '#4ECDC4'],
           text=gender_counts.values, textposition='auto',
           name='Gender', hovertemplate='<b>%{x}</b><br>Count: %{y}<extra></extra>'),
    row=1, col=1
)

# 2. Department GPA - Dynamic colors
dept_colors = px.colors.qualitative.Set3
for i, dept in enumerate(dept_means['Department']):
    fig.add_trace(
        go.Bar(x=[dept], y=[dept_means.loc[i, 'GPA']],
               marker_color=dept_colors[i % len(dept_colors)],
               name=dept, hovertemplate='<b>%{customdata[0]}</b><br>Avg GPA: %{y:.2f}<extra></extra>',
               customdata=[[dept]]),
        row=1, col=2
    )

# 3. GPA Histogram
fig.add_trace(
    go.Histogram(x=df['GPA'], nbinsx=15, name='GPA',
                 marker_color='#A8E6CF', opacity=0.8,
                 hovertemplate='<b>GPA: %{x:.2f}</b><br>Count: %{y}<extra></extra>'),
    row=1, col=3
)

# 4. Scatter Plot with trendline
scatter_fig = px.scatter(df, x='Attendance', y='GPA', trendline="ols",
                         color='Gender', hover_data=['Student_ID'])
for trace in scatter_fig.data:
    fig.add_trace(trace, row=2, col=1)

# 5. Box Plot - GPA by Living
for living_type, color in zip(['Home', 'PG'], ['#FF9999', '#66B2FF']):
    fig.add_trace(
        go.Box(y=df[df['Living']==living_type]['GPA'], name=living_type,
               marker_color=color, hovertemplate=f'<b>{living_type}</b><br>%{{y:.2f}}<extra></extra>'),
        row=2, col=2
    )

# 6. Correlation Heatmap
corr_data = df[['Attendance', 'GPA']].corr()
fig.add_trace(
    go.Heatmap(z=corr_data.values, x=corr_data.columns, y=corr_data.columns,
               colorscale='RdBu_r', zmid=0, hoverongaps=False,
               hovertemplate='<b>%{x}</b> vs <b>%{y}</b><br>Correlation: %{z:.3f}<extra></extra>'),
    row=2, col=3
)

# Layout updates
fig.update_layout(
    title={'text': "Student Performance Analytics Dashboard", 'x': 0.5, 'xanchor': 'center',
           'font': {'size': 24, 'color': '#2C3E50', 'family': 'Arial Black'}},
    height=800, width=1400,
    plot_bgcolor='#F8F9FA',
    paper_bgcolor='#FFFFFF',
    font=dict(size=12, color='#2C3E50'),
    showlegend=True,
    hovermode='closest',
    updatemenus=[{
        "buttons": [
            {"method": "update", "label": "All Departments", "args": [{"visible": [True]*len(fig.data)}]},
            {"method": "update", "label": "Only GPA Charts", "args": [{"visible": [False, True, True, False, True, True]}]}
        ],
        "direction": "down", "showactive": True
    }]
)

fig.update_xaxes(showgrid=True, gridwidth=1, gridcolor='#E9ECEF')
fig.update_yaxes(showgrid=True, gridwidth=1, gridcolor='#E9ECEF')

fig.show()

# 3D scatter plot
print("Creating 3D scatter plot...")
fig_3d = px.scatter_3d(df, x='Attendance', y='GPA', z='Department_Enc',
                      color='Gender', size='GPA', hover_name='Student_ID',
                      title="Attendance vs GPA vs Department",
                      color_discrete_map={'Male': '#FF6B6B', 'Female': '#4ECDC4'})
fig_3d.update_layout(height=700, width=1000)
fig_3d.show()

# Sunburst chart
print("Creating sunburst chart...")
sunburst_data = df.groupby(['Department', 'Living', 'Gender', 'GPA_Category']).size().reset_index(name='count')
sunburst_data = sunburst_data[sunburst_data['count'] > 0]
fig_sunburst = px.sunburst(sunburst_data, path=['Department', 'Living', 'Gender', 'GPA_Category'],
                          values='count',
                          title="Hierarchical Student Distribution",
                          color='count', color_continuous_scale='Viridis')
fig_sunburst.update_layout(height=700, width=800)
fig_sunburst.show()

# Gauge charts
print("Creating performance gauges...")
fig_gauge = make_subplots(1, 3, specs=[[{'type':'indicator'}, {'type':'indicator'}, {'type':'indicator'}]],
                         subplot_titles=('Overall GPA', 'Avg Attendance', 'Top Dept GPA'))

# Overall GPA gauge
fig_gauge.add_trace(
    go.Indicator(mode="gauge+number+delta",
                value=df['GPA'].mean(),
                title={'text': "Overall GPA"},
                delta={'reference': df['GPA'].median()},
                gauge={'axis': {'range': [0, 10]},
                      'bar': {'color': "#00B894"},
                      'threshold': {'line': {'color': "red", 'width': 4},
                                   'thickness': 0.75, 'value': df['GPA'].max()}}),
    row=1, col=1
)

# Attendance gauge
fig_gauge.add_trace(
    go.Indicator(mode="gauge+number",
                value=df['Attendance'].mean(),
                title={'text': "Avg Attendance"},
                gauge={'axis': {'range': [0, 100]},
                      'bar': {'color': "#0984e3"}}),
    row=1, col=2
)

# Top Department GPA gauge
top_dept = dept_means.loc[dept_means['GPA'].idxmax()]
fig_gauge.add_trace(
    go.Indicator(mode="gauge+number",
                value=top_dept['GPA'],
                title={'text': f"Top Dept: {top_dept['Department']}"},
                gauge={'axis': {'range': [0, 10]},
                      'bar': {'color': "#6C5CE7"}}),
    row=1, col=3
)

fig_gauge.update_layout(height=400, title_text="Key Performance Indicators")
fig_gauge.show()

print("\nDashboard complete.")
print("All charts are fully interactive.")
```
